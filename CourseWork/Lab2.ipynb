```python
import numpy as np
X = [[1,1],[1,0],[0,1],[0,0]]
t = [1,-1,-1,-1]
W = [0, 0]
b = 0
theta = 0.2
alpha = 0.5
epoch = 0
while 1:
    Y_pred = []
    for i,x in enumerate(X):
        y_in = np.sum(np.dot(W,x)) + b
        if y_in > theta:
            y = 1
        elif y_in < -theta:
            y = -1
        else:
            y = 0
        Y_pred.append(y)
        if y != t[i]:
            W = W + alpha * t[i] * np.array(x)
            b = b + alpha * t[i]
    Y_pred = np.array(Y_pred)
    print("Epoch=", epoch)
    epoch += 1
    print("Y=", Y_pred, "T=", t)
    if np.array_equal(Y_pred, t):
        break
print("W=", W, "B=", b)
```
