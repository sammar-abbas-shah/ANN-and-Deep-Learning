```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
edges = [('A','B'),('A','C'),('B','D'),('B','E'),('C','F'),('C','G')]
G.add_edges_from(edges)
plt.figure(figsize = (8,6))
plt.axis('off')
pos = nx.spring_layout(G,seed = 0)
nx.draw_networkx(G,pos = pos,node_size = 600,
                 node_color = 'skyblue',font_size = 14,font_color ='white',with_labels=True)
plt.title("Graph Visualization")
plt.show()
```

```python
########################################################################################
all_nodes = list(set([node for edge in edges
                      for node in edge]))
num_nodes = len(all_nodes)
node_to_idx = {node:idx for idx,node in enumerate(all_nodes)}
adj_matrix =  np.zeros((num_nodes,num_nodes))
for src,dst in edges:
    src_idx = node_to_idx[src]
    dst_idx  = node_to_idx[dst]
    adj_matrix[src_idx,dst_idx] = 1
    adj_matrix[dst_idx,src_idx] = 1
```
