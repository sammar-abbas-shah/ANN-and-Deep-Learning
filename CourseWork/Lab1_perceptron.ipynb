```python
import numpy as np
```

```python
X = [[1,1],[1,0],[0,1],[0,0]]
t = [1,-1,-1,-1]

W = [0, 0]
b = 0
alpha = 1
epoch = 0
```

```python
while True:
    epoch += 1
    error_count = 0
    print("\nEpoch", epoch)

    for i in range(len(X)):
        x1, x2 = X[i]
        y = 1 if (W[0]*x1 + W[1]*x2 + b) >= 0 else -1

        if y != t[i]:
            W[0] = W[0] + alpha * t[i] * x1
            W[1] = W[1] + alpha * t[i] * x2
            b = b + alpha * t[i]
            error_count += 1

    print(f"Sample {i + 1}: Input={X[i]}, Target={t[i]}, Pred={y}, Weights={W}, Bias={b}")

    if error_count == 0:
        print("Training complete at epoch", epoch)
        break
```
