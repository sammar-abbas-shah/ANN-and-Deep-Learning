```python
import numpy as np
import matplotlib.pyplot as plt

z = np.arange(-4, 4, 0.05)
print(len(z), z)
```

```python
###################################
def sigmoid_activation(z):
    s = 1.0/(1.0+np.exp(-z))
    d = s*(1-s)
    return s, d

output, derivative = sigmoid_activation(z)

plt.plot(z, output, label='sigmoid')
plt.plot(z, derivative, color='red', label='derivative')
plt.title("Sigmoid Activation (0-1)")
plt.xlabel("z")
plt.ylabel("f(z)")
plt.grid()
plt.legend()
plt.show()
```

```python
##############################
# tanh
def tanh_activation(z):
    t = np.tanh(z)
    d = 1 - np.tanh(z)**2
    return t, d
output, derivative = tanh_activation(z)

plt.plot(z, output, label='tanh')
plt.plot(z, derivative, color='red', label='Derivative')
plt.title("Tanh Activation (-1 to 1)")
plt.xlabel("z")
plt.ylabel("f(z)")
plt.grid()
plt.legend()
plt.show()
```

```python
##############################
def ReLU_activation(z):
    output = []
    derivative = []
    for x in z:
        if x < 0:
            output.append(0)
            derivative.append(0)
        else:
            output.append(x)
            derivative.append(1)
    return np.array(output), np.array(derivative)

output, derivative = ReLU_activation(z)

plt.plot(z, output, label='ReLU')
plt.plot(z, derivative, color='red', label='Derivative')
plt.title("ReLU Activation (0 to +∞)")
plt.xlabel("z")
plt.ylabel("f(z)")
plt.grid()
plt.legend()
plt.show()
```

```python
##############################
# Softmax
##############################
z = np.array([4,2,5,0,3,1])

def softmax_activation(z):
    exponents = np.exp(z - np.max(z))
    return exponents / np.sum(exponents)

p = softmax_activation(z)
print("Probabilities = ", p)


#install these libraries
#pillow
#matplotlib
#opencv-python
#scikit- image
#tensorflow
```
