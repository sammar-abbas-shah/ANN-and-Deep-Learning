```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
```

```python
audio, sr = librosa.load(r"D:\ANN and DL\53550.mp3", sr=None)
print('sample rate:', sr)
print('audio shape:', audio.shape)

plt.figure(figsize=(12,4))
librosa.display.waveshow(audio, sr=sr)
plt.title("Waveform of audio sample")
plt.tight_layout()
plt.show()
```

```python
audio, sr = librosa.load(r"D:\ANN and DL\53550.mp3", sr=16000)
mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=16000)
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%2.0f dB')
plt.title('Mel Spectrogram')
plt.tight_layout()
plt.show()
```

```python
chromagram = librosa.feature.chroma_stft(y=audio, sr=16000)
chromagram_db = librosa.power_to_db(chromagram, ref=np.max)

frames = range(chromagram.shape[1])
times = librosa.frames_to_time(frames, sr=sr)
######################################################
plt.figure(figsize=(10, 4))
librosa.display.specshow(chromagram_db, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title("Chromagram")
plt.tight_layout()
plt.show()
```
