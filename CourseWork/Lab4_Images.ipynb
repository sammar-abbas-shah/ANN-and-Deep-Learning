```
###################################
import cv2
def display_image(text,i):
    cv2.imshow(text,i)
    cv2.waitKey(0)
```

```
##################################
image1 = cv2.imread('r.webp')
image0 = image1.copy()
display_image('Original Image ',image1)
image2 = cv2.cvtColor(image0,cv2.COLOR_BGR2RGB)
display_image('RGB Image ',image2)
image3 = cv2.resize(image1,(100,120))
#display_image(image3)
display_image('Resized Image', image3)

image_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
display_image('Gray Image', image_gray)
```
