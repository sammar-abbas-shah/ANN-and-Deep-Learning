```python

import gymnasium as gym
env = gym.make('FrozenLake-v1',render_mode='human')
observation,info = env.reset()
print('Observation',observation)
print('Info',info)
for i in range(50):
    action = env.action_space.sample()
    observation,reward,terminated,truncated,info = env.step(action) # returns 5 variables  terminated means agent reached goal truncate agent fall into hole
    if terminated or truncated:
        observation,info = env.reset()
env.close()
```
